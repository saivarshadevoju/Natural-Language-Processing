## Named Entity Recognition

```python
sentence = "The Eiffel Tower was built from 1887 to 1889 by French engineer Gustave Eiffel, whose company specialized in building metak frameworks and structures"
```

```python
import nltk
tags = nltk.pos_tag(nltk.word_tokenize(sentence))
print(tags)
```

```python
nltk.download('maxent_ne_chunker')
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')
nltk.ne_chunk(tags)
```

```python
nltk.ne_chunk(tags).draw()
```

```python
sen = "I Love India"
tag = nltk.pos_tag(nltk.word_tokenize(sen))
nltk.ne_chunk(tag)
```
