```python
corpus = """My name is varsha! I love cooking various dishes. 
I am also a Malla Reddy's student.
"""
```

```python
print(corpus)
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
#Tokenization 
# Paragraph ---> Sentences

from nltk.tokenize import sent_tokenize
documents = sent_tokenize(corpus)
print(documents)
```

```python
for sentence in documents:
    print(sentence)
```

```python
#Tokenization
# Paragraph ---> Words
# Sentences ---> Words

from nltk.tokenize import word_tokenize
words = word_tokenize(corpus)
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```
